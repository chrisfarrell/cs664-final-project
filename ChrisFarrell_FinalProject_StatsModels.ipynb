```python
import statsmodels.api as sm
import pandas
from patsy import dmatrices
```

## load data from raw.github.com/vincentarelbundock/Rdatasets/master/csv/pscl/presidentialElections.csv

```python
df = sm.datasets.get_rdataset("presidentialElections", "pscl").data
```

## Inspect tail

```python
df[-5:]
```

## clean up: drop any rows with missing data

```python
df = df.dropna()
df[-5:]
```

## derive design matrices

```python
y, X = dmatrices('demVote ~ state + year + south', data=df, return_type='dataframe')
```

## inspect the matrix of dependent variables (y)

```python
y[:3]
```

## inspect the matrix of predictor variables (X)

```python
X[:3]
```

## estimate the model using Ordinary Least Squares regression

```python
mod = sm.OLS(y, X)
```

## fit the model

```python
res = mod.fit()
```

## inspect the results

```python
print res.summary()
```

## extract parameter estimates

```python
res.params
```

## R squared

```python
res.rsquared
```

## Rainbow test for linearity: (F-stat, p-value)

```python
sm.stats.linear_rainbow(res)
```

## Partial Regression Plot

```python
sm.graphics.plot_partregress('demVote', 'year', ['south', 'state'], 
                             data=df, obs_labels=False)
```
