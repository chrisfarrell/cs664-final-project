## load the presidential csv file and make a data frame

```python
file_presidential_elections = open('elections-1900-to-2008.csv')
frame_all_candidates = pd.read_csv(file_presidential_elections, header=None, index_col=0)
```

## fill missing values

```python
frame_all_candidates = frame_all_candidates.fillna(method='ffill', axis=1)
```

## drop uneeded columns

```python
del_cols = np.arange(3,84,3)
frame_all_candidates.drop(del_cols,inplace=True,axis=1)
```

## edit multi index columns

```python
frame_all_candidates.columns = pd.MultiIndex.from_arrays(frame_all_candidates[:2].values, names=['year','candidate'])
frame_all_candidates.index.name = 'state'
frame_all_candidates = frame_all_candidates.drop('State')
frame_all_candidates = frame_all_candidates[pd.notnull(frame_all_candidates.index)]
```

## fill empty cells with '0'

```python
frame_all_candidates.replace(to_replace=' ', value=0, inplace=True)
```

## convert all cells to floats so we can do math and plot

```python
frame_all_candidates = frame_all_candidates[:][:].astype(float)
```

## slice data into democrats and republicans

```python
frame_all_democrats = frame_all_candidates.iloc[:,::2]
frame_all_republicans = frame_all_candidates.iloc[:,1::2]
```

## machine learning: Classification, K-th Nearest Neighbor

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_X_data = frame_all_democrats.iloc[:,1:28].values
knn_Y_target = frame_all_democrats.iloc[:,:1].values
```

```python
frame_all_democrats.iloc[:,1:28]
```

```python
frame_all_democrats.iloc[:,:1]
```

```python
np.random.seed(0)
knn_indices = np.random.permutation(len(knn_X_data))
```

```python
knn_X_train = knn_X_data[knn_indices[:-10]]
knn_Y_train = knn_Y_target[knn_indices[:-10]].ravel()
```

```python
print knn_X_train
```

```python
print knn_Y_train
```

```python
knn_X_test = knn_X_data[knn_indices[-10:]]
knn_Y_test = knn_Y_target[knn_indices[-10:]].ravel()
```

```python
print knn_X_test
```

```python
knn = KNeighborsClassifier()
knn.fit(knn_X_train, knn_Y_train)
knn_prediction = knn.predict(knn_X_test)
```

```python
print "prediction: " , knn_prediction.astype(int)
print "test data:  " , knn_Y_test.astype(int)
```

```python
# The mean square error
print("Residual sum of squares: %.2f"
      % np.mean((knn_prediction - knn_Y_test) ** 2))

# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % knn.score(knn_X_test, knn_Y_test))
```

```python
plt.scatter(knn_Y_test.astype(int), knn_prediction.astype(int))
```

## machine learning: Regression, Lasso

```python
from sklearn import linear_model
```

```python
clf_X_data = frame_all_democrats.iloc[:,1:28].values
clf_Y_target = frame_all_democrats.iloc[:,:1].values
clf = linear_model.Lasso(alpha = 0.015, tol=0.0001, max_iter=100000, fit_intercept=False, positive=True)
clf_X_train = clf_X_data[:-10]
clf_Y_train = clf_Y_target[:-10].ravel()
clf_X_test = clf_X_data[-10:]
clf_Y_test = clf_Y_target[-10:].ravel()
clf.fit(clf_X_train, clf_Y_train)
clf_prediction = clf.predict(clf_X_test).astype(int)
print "prediction: " , clf_prediction.astype(int)
print "test data:  " , clf_Y_test.astype(int)

```

```python
# The mean square error
print("Residual sum of squares: %.2f"
      % np.mean((clf_prediction - clf_Y_test) ** 2))

# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % clf.score(clf_X_test, clf_Y_test))
```

```python
plt.scatter(clf_Y_test.astype(int), clf_prediction.astype(int))
```
